## Playing Around with Clustering Days into Different Groups

```python
cluster_vars = [
    'Steps',
    'Screen_Time_min',
    'habit_count_noworkout',
    'workout',
    'Meditation',
    'Calories(kcal)',
]

df_cluster = data_bos[cluster_vars].dropna()

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 8), inertia, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df_cluster['day_type'] = clusters
```

```python
cluster_profiles = (
    df_cluster
    .groupby('day_type')
    .mean()
)

cluster_profiles
```

```python
cluster_profiles_unscaled = pd.DataFrame(
    scaler.inverse_transform(cluster_profiles),
    columns=cluster_vars,
    index=cluster_profiles.index
)
```

```python
cluster_profiles_unscaled.head()
```

```python
data_bos = data_bos.merge(
    df_cluster[['day_type']],
    left_index=True,
    right_index=True,
    how='left'
)
```

```python
weekday_cluster = (
    data_bos
    .groupby(['DayofWeek', 'day_type'])
    .size()
    .unstack(fill_value=0)
)

weekday_cluster_pct = weekday_cluster.div(
    weekday_cluster.sum(axis=1), axis=0
)

weekday_cluster_pct
```

```python
weekday_order = [
    'Monday', 'Tuesday', 'Wednesday',
    'Thursday', 'Friday', 'Saturday', 'Sunday'
]

weekday_cluster_pct = weekday_cluster_pct.loc[
    weekday_cluster_pct.index.intersection(weekday_order)
]
weekday_cluster_pct = weekday_cluster_pct.reindex(weekday_order)
```

```python
import matplotlib.pyplot as plt

ax = weekday_cluster_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(9, 5),
    edgecolor='black'
)

ax.set_ylabel("Proportion of days")
ax.set_xlabel("")
ax.set_title("Distribution of Behavioral Day Types by Weekday")

# Move legend outside
ax.legend(
    title="Day Type",
    bbox_to_anchor=(1.02, 1),
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

## Priorities of Different Days

```python

# Make sure weekday_order is consistent
#weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Count how many times each Priority_AM category appears per weekday

#group by day of the week and category, then counts the number of rows in each group
priority_counts = data_bos.groupby(['DayofWeek', 'Priority_AM']).size().unstack(fill_value=0).reindex(weekday_order)
#sum across each weekday, divide each value by the row sum to convert to proportions
priority_counts_norm = priority_counts.div(priority_counts.sum(axis=1), axis=0)

#plot
ax=priority_counts_norm.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')

#add labels 
for i, weekday in enumerate(priority_counts_norm.index):
    bottom=0
    for category in priority_counts_norm.columns:
        height=priority_counts_norm.loc[weekday,category]
        if height > 0:
            ax.text(
            i,
            bottom + height/2,
            f"{height:.2f}",
            ha='center', va='center', fontsize=11, color='black'
            )
            bottom+= height

plt.title("AM Priority Categories Across Week", fontsize=14)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Proportion of Responses", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Priority_AM', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
outfile=os.path.join(save_dir,"Priorities_Bar.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')
plt.show()
```

## Radial Plot for Habit Completion

```python
import numpy as np
import matplotlib.pyplot as plt

# --- SETTINGS ---

habits_list=['Clean_5min', 'Journal', 'workout', 'consume_literature',
       'Turkish', 'Creatine', 'Meditation']
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# --- COMPUTE MEANS ---
weekly_means = (
    data_bos.groupby('DayofWeek')[habits_list]
            .mean()
            .reindex(weekday_order)
)

# --- NORMALIZE each metric to 01 for radial comparability ---
normalized = (weekly_means - weekly_means.min()) / (weekly_means.max() - weekly_means.min())

# --- RADAR SETUP ---
labels = habits_list
num_vars = len(labels)

# Angles for radar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the loop

# Setup figure
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot each weekday
for day in weekday_order:
    values = normalized.loc[day].tolist()
    values += values[:1]  # close the loop
    
    ax.plot(angles, values, linewidth=2, label=day)
    ax.fill(angles, values, alpha=0.1)

# --- STYLING ---
ax.set_title("Habit Completion by Weekday (Radar Chart)", fontsize=18, pad=20)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)
ax.set_yticklabels([])  # hide radial labels for cleanliness

plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
outfile=os.path.join(save_dir,"RadarChart.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')

plt.show()
```

## Habit Fragility -- probably copy to the habit notebook

```python
vars_fragility = [
    #'Steps',
    'Screen_Time_min',
    'Clean_5min', 
    'Journal', 
    'workout', 
    'consume_literature',
       'Turkish', 
    'Creatine', 
    'Meditation',    
    #'Calories(kcal)',
    'PickUps',
    'workout'
]
```

```python
weekday_stats = (
    data_bos
    .groupby('DayofWeek')[vars_fragility]
    .agg(['mean', 'std'])
)
```

```python
weekday_means = weekday_stats.xs('mean', axis=1, level=1)
weekday_std   = weekday_stats.xs('std', axis=1, level=1)
```

```python
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekday_means = weekday_means.loc[weekday_order]
weekday_std   = weekday_std.loc[weekday_order]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

for var in vars_fragility:
    plt.scatter(
        weekday_means[var],
        weekday_std[var],
        label=var,
        s=80
    )

plt.xlabel("Weekday Mean")
plt.ylabel("Weekday Variability (SD)")
plt.title("Behavioral Fragility Across Weekdays")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
weekday_cv = weekday_std / weekday_means
```

```python
plt.figure(figsize=(8, 6))

for var in vars_fragility:
    plt.scatter(
        weekday_means[var],
        weekday_cv[var],
        label=var,
        s=80
    )

plt.xlabel("Weekday Mean")
plt.ylabel("Coefficient of Variation (Fragility)")
plt.title("Behavioral Fragility (Scale-Free)")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
fragility_rank = weekday_cv.mean().sort_values(ascending=False)
fragility_rank
```

## Weekly Behavioral Groups

```python
activity_vars=['HR_max', 'workout', 'Steps','Calories(kcal)']
screen_time_vars=["Screen_Time_min", 'Message_notifications', "Instagram", 'TikTok', 'PickUps']
habits_vars='Clean_5min', 'Journal', 'consume_literature', 'Turkish', 'Creatine', 'Meditation'
```

```python
# Variables
activity_vars = ['HR_max', 'workout', 'Steps','Calories(kcal)']
# screen_time_vars = ["Screen_Time_min", 'Message_notifications', "Instagram", 'TikTok', 'PickUps']
screen_time_vars = ["Screen_Time_min", 'Message_notifications', 'PickUps']
habits_vars = ['Clean_5min', 'Journal', 'consume_literature', 'Turkish', 'Creatine', 'Meditation']
Mental_PM= ['Mental_PM']

# Group variables by type
var_groups = {
    'Activity': activity_vars,
    'Screen Time': screen_time_vars,
    'Habits': habits_vars,
    'Mental_PM': Mental_PM
}

# Compute weekday mean per variable
weekday_means = data_bos.groupby('DayofWeek')[activity_vars + screen_time_vars + habits_vars+Mental_PM].mean()

# Reorder weekdays
weekday_order = ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday']
weekday_means = weekday_means.loc[weekday_order]

# Optional: Z-score per variable for heatmap comparability
weekday_z = (weekday_means - weekday_means.mean()) / weekday_means.std()
```

```python
weekday_grouped = pd.DataFrame(index=weekday_z.index)

for group_name, vars_list in var_groups.items():
    weekday_grouped[group_name] = weekday_z[vars_list].mean(axis=1)


weekday_grouped_sorted=weekday_grouped.sort_values(by='Mental_PM',ascending=False)

weekday_grouped_sorted["Behavioral_Load"] = (
    weekday_grouped_sorted[['Activity', 'Screen Time', 'Habits']]
    .mean(axis=1)
)

weekday_grouped_sorted=weekday_grouped_sorted[
    ['Activity', 'Screen Time', 'Habits', 'Behavioral_Load', 'Mental_PM']
]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(weekday_grouped_sorted, annot=True, cmap='RdYlGn', center=0, linewidths=0.5)
plt.axvline(x=4, color='black', linewidth=3)
plt.axvline(x=3, color='white', linewidth=2)

plt.title("Weekly Behavioral Groups (Z-scored)")
plt.ylabel("Weekday")
plt.xlabel("Behavioral Group")
outfile=os.path.join(save_dir,"Heatmap.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')

plt.show()
```
