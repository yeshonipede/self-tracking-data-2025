## Different Physio Metrics Across the Week

```python
weekly_means = (
    data_bos.groupby('DayofWeek')[physio_metrics]
            .mean()
            .reindex(weekday_order)
)

physio_metrics=['Energy_AM','Fatigue_PM','TimeAsleep_min', 'HRV_avg', 'HR_min']
labels=["My AM Energy Rating", "My PM Fatigue Rating", "Time Asleep (min)", "Average HRV (bpm)", "Resting HR (bpm)"]

fig, axes = plt.subplots(len(physio_metrics), 1, figsize=(10, 12), sharex=True)
for i, col in enumerate(physio_metrics):
    axes[i].plot(weekly_means.index, weekly_means[col], marker='o', color='C'+str(i))
    axes[i].set_ylabel(labels[i], fontsize=10)
    axes[i].grid(True)
axes[-1].set_xlabel("Weekday")

outfile=os.path.join(save_dir,"Fatigue.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')
plt.show()
```

```python
#Creating columns for AM Morale and PM Morale 
data_bos['AM_morale']=data_bos[["Optimism_AM", "Mind_AM"]].mean(axis=1)
data_bos['PM_morale']=data_bos[["Mental_PM","Satisfied_PM", "Fun_PM", "Intention_Satisfaction_PM"]].mean(axis=1)
```

## Priorities Across the Week

```python

# Make sure weekday_order is consistent
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Count how many times each Priority_AM category appears per weekday

#group by day of the week and category, then counts the number of rows in each group
priority_counts = data_bos.groupby(['DayofWeek', 'Priority_AM']).size().unstack(fill_value=0).reindex(weekday_order)
#sum across each weekday, divide each value by the row sum to convert to proportions
priority_counts_norm = priority_counts.div(priority_counts.sum(axis=1), axis=0)

#plot
ax=priority_counts_norm.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')

#add labels 
for i, weekday in enumerate(priority_counts_norm.index):
    bottom=0
    for category in priority_counts_norm.columns:
        height=priority_counts_norm.loc[weekday,category]
        if height > 0:
            ax.text(
            i,
            bottom + height/2,
            f"{height:.2f}",
            ha='center', va='center', fontsize=11, color='black'
            )
            bottom+= height

plt.title("AM Priority Categories by Weekday", fontsize=14)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Proportion of Responses", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Priority_AM', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
outfile=os.path.join(save_dir,"Priorities_Bar.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')
plt.show()
```

## AM and PM Morale Throughout the Week

```python
weekly_means_AM=data_bos.groupby('DayofWeek')['AM_morale'].mean().reindex(weekday_order)
weekly_means_PM=data_bos.groupby('DayofWeek')['PM_morale'].mean().reindex(weekday_order)

#combine into one df
weekly_scores=weekly_means_AM.to_frame('AM_morale').join(weekly_means_PM.to_frame('PM_morale'))

#Plot details 
plt.figure(figsize=(10,6))
plt.plot(weekly_scores.index, weekly_scores['AM_morale'],marker='o', color='orange', linewidth=2, label="AM Morale")
plt.plot(weekly_scores.index, weekly_scores['PM_morale'], marker='s', color='navy', linewidth=2, label="PM Morale")

plt.title('Average Self-Reported Scores by Day of the Week')
plt.xlabel('Weekday')
plt.ylabel('Average Morale Score Based on Self-Reported Data')
plt.xticks(rotation=45)
plt.legend(title='Form')
plt.tight_layout()
plt.grid(True)
outfile=os.path.join(save_dir,"Morale.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')
plt.show()
```

## Prediction Errors Throughout the Week

```python
# Create prediction error columns
data_bos['Satisfaction_Error'] = data_bos['How_Well_Prediction_AM'] - data_bos['Satisfied_PM']
data_bos['Crazy_Error'] = data_bos['Crazy_Prediction_AM'] - data_bos['Crazy_PM']

# Compute mean error per weekday
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
weekly_error = data_bos.groupby('DayofWeek')[['Satisfaction_Error', 'Crazy_Error']].mean().reindex(weekday_order)
weekly_error
```

```python
plt.figure(figsize=(10,6))
plt.plot(weekly_error.index, weekly_error['Satisfaction_Error'], marker='o', color='lightseagreen', label='Satisfaction Error')
plt.plot(weekly_error.index, weekly_error['Crazy_Error'], marker='s', color='mediumvioletred', label='Crazy Error')

plt.axhline(0, color='k', linestyle='--')  # zero reference line
plt.title("Weekly Prediction Errors")
plt.xlabel("Day of the Week")
plt.ylabel("Prediction Error (AM - PM)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
outfile=os.path.join(save_dir,"PredictionErrors.png")
plt.savefig(outfile,dpi=300,bbox_inches='tight')
plt.show()
```

## Radial Plot for Habit Completion

```python
import numpy as np
import matplotlib.pyplot as plt

# --- SETTINGS ---

habits=['Clean_5min', 'Journal', 'workout', 'consume_literature',
       'Turkish', 'Creatine']
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# --- COMPUTE MEANS ---
weekly_means = (
    data_bos.groupby('DayofWeek')[habits]
            .mean()
            .reindex(weekday_order)
)

# --- NORMALIZE each metric to 01 for radial comparability ---
normalized = (weekly_means - weekly_means.min()) / (weekly_means.max() - weekly_means.min())

# --- RADAR SETUP ---
labels = habits
num_vars = len(labels)

# Angles for radar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # close the loop

# Setup figure
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot each weekday
for day in weekday_order:
    values = normalized.loc[day].tolist()
    values += values[:1]  # close the loop
    
    ax.plot(angles, values, linewidth=2, label=day)
    ax.fill(angles, values, alpha=0.1)

# --- STYLING ---
ax.set_title("Habit Completion by Weekday (Radar Chart)", fontsize=18, pad=20)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)
ax.set_yticklabels([])  # hide radial labels for cleanliness

plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

plt.show()
```
