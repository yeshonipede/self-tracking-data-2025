```python
def normalize_all_dates(dfs_dict):
    for name, df in dfs_dict.items():
        if 'Date' in df.columns:
            df['Date'] = pd.to_datetime(
                df['Date'],
                format='mixed',
                errors='raise'
            ).dt.normalize()
```

```python
def convert_to_min(value):
    if pd.isna(value):
        return None

    valu = str(value).lower()  # handle HR, Min, etc.
    hours = 0
    minutes = 0

    # Match hours (h, hr)
    h_match = re.search(r'(\d+)\s*(h|hr)', valu)
    if h_match:
        hours = int(h_match.group(1))

    # Match minutes (m, min, mins)
    m_match = re.search(r'(\d+)\s*(m|min|mins)', valu)
    if m_match:
        minutes = int(m_match.group(1))

    return hours * 60 + minutes
```

```python
#Converts columns in the form hours, m to being exclusively in minutes
dfs = [screen_time, media_consumed, med_month]
cols=["Screen_Time", "Audiobooks", "Minutes"]
for df in dfs:
    for col in cols:
        if col in df.columns:
            df[col + '_min'] = df[col].apply(convert_to_min)
        
```
